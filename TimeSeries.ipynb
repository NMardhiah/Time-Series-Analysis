#**DATA CLEANING**


---



```python
#While checking on the columns we have observed that uncessesary charaters which  
#need to amend by converting the columns name into list for easy processing 

list=wheat.columns.to_list()
print(list)
```

```python

#we notices that there is a columns named country and other year columns in the dataset

clean_list=[]
for column in list:

#if there is country column then need to append that
    if column== 'Country':
        column=='Country'
        clean_list.append(column)

#for all other columns where cleaning is required take only four digits 'year'
    else:
        clean_column=column[0:4] #appending in the clean_list(cleaned list of the column)
        clean_list.append(clean_column)

print(clean_list)
```

```python
#Now, we can renaming the colums with 4 digit year
wheat.set_axis (['Country', '2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', 
                 '2012', '2011', '2010', '2009', '2008', '2007', '2006', '2005', '2004',
                  '2003', '2002', '2001', '2000', '1999', '1998', '1997', '1996'],axis='columns',
                inplace=True)

wheat.columns #to call the column country that contained year 
wheat.head(3) #to call top 3 data
```

```python
#To make the table more easy to read, we will make one new column that 
#contained all the values with different years column by using melt function in Pandas

variables= ['2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', 
         '2012', '2011', '2010', '2009', '2008', '2007', '2006', '2005', '2004',
         '2003', '2002', '2001', '2000', '1999', '1998', '1997', '1996']

#.melt:Unpivot a DataFrame from wide to long format / ?wheat.melt
#id_vars (identifier variables), value_vars(measured variables)

wheat_new=wheat.melt(id_vars=['Country'], value_vars=variables, var_name='Year', 
                     value_name='Value')
wheat_new.head(100)
```

```python
#As we do on checking towards the data, we noticed in the row 95 which the data
#showed something like 14.5[7], which this will be the problem to perform
#analysis over this kind of the data
#hence, we need to remove the uncessecary at the end of the number

#first, before cleaning, number/value need to convert the column in list format
value_list=wheat_new.Value.to_list()

#second, to clean the data, we will keep the number and remove the character
# after '[' at the end
cleaned_value_list=[]

for value in value_list:
    clean_value=str(value).split('[')[0] 
    cleaned_value_list.append(clean_value)

#third, we assign the clean value to the column
wheat_new.Value=cleaned_value_list

wheat_new.head(100)
```

```python
#Checking the information on the new column 
wheat_new.info()
```

```python
#We need to convert the data type for 'Value' from 'object' to 'float64'
#in order to perform an arithmetic operation

wheat_new.Value=wheat_new.Value.astype('float64')
wheat_new.info()
```

```python
# brief statistical description about for the value 
wheat_new.describe()
```

##**1. Top 10 years when the wheat production is the highest**


> Below table showed the top 10 years of the wheat production is the highest,

```python
#Next, we will checking the wheat production on the year wise
#In this case, we will groupby function used to group the year column and sum the values 

wheat_year=wheat_new.groupby(['Year','Country'])['Value'].sum().reset_index().sort_values(by='Value',
                             ascending=False)

wheat_year_wise_countries=wheat_year.loc[wheat_year.Country!='World total'] #not including world total

wheat_year_wise_countries
```

##**2. Maximum production across all the countries**

```python
#creating an empty list
year=[]

#to group the country based on the maximum wheat production 
new_df=pd.DataFrame(wheat_new.groupby('Country')['Value'].max())

for v,name in zip(new_df.Value , new_df.index):

  y=wheat_new.loc[(wheat_new['Country']==name) & (wheat_new['Value']==v),'Year'].values[0]
  year.append(y)

print(year)
```

```python
#to call the new function as new_df 
wheat_countries_wise_year=new_df
wheat_countries_wise_year['Max year']=year

wheat_countries_wise_year.sort_values(by='Value',ascending=False).head(10)
                    
```

##**4. Wheat Production Statistics by Year**

```python
#set the country code from the country names for easy to search
#!pip install pycountry
#!pip install geopandas

import pycountry
import geopandas as gpd
```

```python
#to make an extra things to add a new attribtes to the dataset 
def country_code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE

wheat['iso_a3']=country_code(wheat.Country)
wheat.head()
   
```

##**5. Wheat Production Separated by Region(Hemisphere)**


*   Showing the most popular region that produce the wheat according to the hemispere




```python
#classified the country to classified to their region(hemisphere) 

hemisphere = ["North", "North", "North", "North", "North",
             "North", "North", "North", "North", "North",
             "South", "North", "South", "North", "North East",
             "North", "North", "West", "North", "North", 
             "South", "North", "North East", "South", "North East", 
             "North East", "North", "North East", "North East", "East",
             "North", "North", "North West", "North East", "North East", "North", "North",
             "North", "North", "North", "South", "Total"]

wheat["region"] = hemisphere 
```

```python
#calling the hemispere
hemisphere [0:10]
```

```python
#??matplotlib.text.Text
```

```python
#to plot the pie chart for the hemispere 

fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
regionsCount = wheat.region.value_counts()[:-1] #remove the world total

#For pie plotting design 
palette = ["#F2B8F5", "#e4c2ac", "#9DDA6D", "#F5A2D3", "#A2CEF5", "#A2ABF5"]
wedgeprops = {"edgecolor":"brown",
              "width":.5}
textprops = {"fontsize":15}

#args-non keywords argument, kwargs - keyword arguments
ax[0].bar(regionsCount.index, regionsCount.values, color=palette, hatch="..")
ax[0].set_ylabel("Count")
ax[0].set_xlabel("Hemispere")
for region,count in zip(regionsCount.index, regionsCount.values):
    ax[0].annotate(f"{count}",
                   xy=(region,count+.4),
                   va="center", ha="center", fontweight="bold", fontsize=9, color="red")
ax[0].set_ylim([0,27])

_, _, autopcts = ax[1].pie(x=regionsCount.values, labels=regionsCount.index, 
                           colors=palette, shadow=True, autopct="%1.1f%%", 
                           explode=[.1,0,0,0,0,0], wedgeprops=wedgeprops, textprops=textprops, 
                           pctdistance=1.15, labeldistance=1.3)
plt.setp(autopcts, **{"fontweight":"bold", "fontsize":9, "color":"red"})

fig.text(.125,1,"Wheat Production by Regions", fontsize=15, fontweight="bold")
fig.text(.125,.9325,"Northern regions in every coutry are typically the most popular \nlocation for wheat plantation amongst wheat producers.", fontsize=12, fontweight="medium")
plt.show()
```


> From the pie chart , we can identify which region that produce the most wheat production which is North overall by 61% and followed by North East by 19.5%. We can assume that this is due to the soil of the land which it is more suitable to plant the wheat rather than other region. 



##**6. Wheat Production Separated by Continents**


*   Which country that contribute the most according to the continents




```python
#classified the country to classified to their continents

continents = ["Asia", "Asia", "Eurasia", "North America", "North America", "Europe", "Asia", "Europe", "Europe", "Eurasia", "South America",
             "Asia", "Oceania", "Eurasia", "Europe", "Europe", "North Africa", "Europe", "Europe", "Europe", "South America", "Asia", "Asia",
             "North Africa", "Asia", "Europe", "Europe", "Europe", "Europe", "Europe", "Europe", "North Africa", "North America", "Europe", "Asia",
             "Europe", "Europe", "North Africa", "Asia", "Europe", "South Africa", "Total"]

wheat["continent"] = continents
continents [0:10]
```

```python
#plot  a pie chart based on region 

fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(15,6))
continentsCount = wheat.continent.value_counts().sort_values(ascending=True).reindex(["Total","Oceania","South Africa","South America",'Eurasia',
                                                                  'North America', 'North Africa', 'Asia', 'Europe'])[1:]

# For pie plotting design
palette = ["#F5CCAA", "#e4c2ac", "#9EC43D", "#A4F3EA", "#6EE7BF", "#ABC8CB", "#CCEF7B","#AAC9F5"]
wedgeprops = {"edgecolor":"black",
              "width":.5}
textprops = {"fontsize":15}


ax[0].barh(continentsCount.index, continentsCount.values, color=palette, hatch="///")
for count,continent in zip(continentsCount.values, continentsCount.index):
    ax[0].annotate(f"{count}",
                   xy=(count+.5,continent),
                   va="center", ha="center", fontweight="bold", fontsize=10, color="darkgreen")
ax[0].set_xlim([0,20])
ax[0].spines["left"].set_visible(True)
ax[0].set_xlabel("Count") #41 total of countries , except world total
ax[0].set_ylabel("Region") 

_, _, autopcts = ax[1].pie(x=continentsCount.values, labels=continentsCount.index, colors=palette, 
                           shadow=True, autopct="%1.1f%%", 
                           explode=[0,0,0,0,0,0,.1,.1], wedgeprops=wedgeprops, textprops=textprops, 
                           pctdistance=1.15, labeldistance=1.3,
                           startangle=-45)
plt.setp(autopcts, **{"fontweight":"bold", "fontsize":8, "color":"blue"})

fig.text(.125,.96,"Wheat Producer by Continents", fontsize=15, fontweight="bold", color="#823617")
fig.text(.125,.89,"Majority of the wheat producers are European or Asian countries,\nwith a few from America, Africa, and Oceania.", 
         fontsize=12, fontweight="medium", color="saddleBrown")
plt.show()
```

From the graph, we can see that which contnents that produce the most of the wheat which are Eurasia's (Russia) by 43.9% ,followed by Asia (China) by 134.3 million tonnes in total and North Africa (Egypt) by 9.8%. Majority of the wheat producers are from European or Asian countries with a few from America, Africa and Ocenia.
